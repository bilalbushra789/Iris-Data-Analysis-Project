## Dataset Description
The **Iris Dataset** contains 150 samples of iris flowers from 3 species:
- *Iris Setosa*
- *Iris Versicolor*
- *Iris Virginica*

Each sample has 4 features:
- **sepal_length** — length of the sepal (cm)
- **sepal_width**  — width of the sepal (cm)
- **petal_length** — length of the petal (cm)
- **petal_width**  — width of the petal (cm)

## Step 5: Visualizations

### 5.1 — Scatter Plot: Relationship Between Variables
Scatter plots help us see how two numeric features relate to each other and whether different species cluster together.

```python
# ─── Scatter Plot 1: Sepal Length vs Sepal Width ─────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Sepal dimensions
sns.scatterplot(
    data=df,
    x='SepalLengthCm',     # Updated to match CSV exact name
    y='SepalWidthCm',      # Updated to match CSV exact name
    hue='Species',         # Capitalized 'S'
    style='Species',       # Capitalized 'S'
    s=80,
    ax=axes[0]
)
axes[0].set_title('Sepal Length vs Sepal Width', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Sepal Length (cm)')
axes[0].set_ylabel('Sepal Width (cm)')

# Plot 2: Petal dimensions — usually more separable between species
sns.scatterplot(
    data=df,
    x='PetalLengthCm',     # Updated to match CSV exact name
    y='PetalWidthCm',      # Updated to match CSV exact name
    hue='Species',         # Capitalized 'S'
    style='Species',       # Capitalized 'S'
    s=80,
    ax=axes[1]
)
axes[1].set_title('Petal Length vs Petal Width', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Petal Length (cm)')
axes[1].set_ylabel('Petal Width (cm)')

plt.suptitle('Scatter Plots — Feature Relationships by Species', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('scatter_plots.png', bbox_inches='tight', dpi=120)
plt.show()

print("Observation: Petal dimensions clearly separate the three species better than sepal dimensions.")
```

```python
# ─── Pair Plot: All Feature Combinations ─────────────────────────────────────
# A pair plot shows scatter plots for every pair of features — great for overview

pair_plot = sns.pairplot(
    df,
    hue='Species',          # Color by species
    diag_kind='kde',        # Diagonal: kernel density estimate
    plot_kws={'alpha': 0.7, 's': 50},
    height=2.5
)
pair_plot.fig.suptitle('Pair Plot — All Feature Combinations by Species', y=1.02, fontsize=14, fontweight='bold')
plt.savefig('pair_plot.png', bbox_inches='tight', dpi=100)
plt.show()
print(" Observation: Petal length & petal width together offer the clearest species separation.")
```

### 5.2 — Histogram: Distribution of Each Feature

```python
# --- Histograms for All 4 Numeric Features ---
# Histograms show how values are distributed for each species

# Updated features list to match your CSV column names
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten() # Flatten 2D array to 1D for easy iteration

for i, feature in enumerate(features):
    # Plot histogram for each species on the same axis
    for spec in df['Species'].unique():
        # Changed 'Species' to 'spec' to match the loop variable
        subset = df[df['Species'] == spec][feature]
        axes[i].hist(
            subset,
            bins=15,
            alpha=0.6,          # Transparency to see overlapping species
            label=spec,
            edgecolor='white'
        )

    # Cleaning up labels (removing 'Cm' and adding space for titles)
    clean_title = feature.replace("Cm", "").replace("_", " ")
    axes[i].set_title(f'Distribution of {clean_title}', fontweight='bold')
    axes[i].set_xlabel(f'{clean_title} (cm)')
    axes[i].set_ylabel('Frequency')
    axes[i].legend(title='Species')

plt.suptitle('Histograms — Feature Distributions by Species', fontsize=15, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('histograms.png', bbox_inches='tight', dpi=120)
plt.show()

print("Observation: Setosa has distinctly smaller petal dimensions – it's easily separated from the others.")
```

```python
# --- KDE (Kernel Density Estimate) Plot — Smoother Distribution View ---
# KDE is like a smoothed histogram — easier to compare shapes across species

# 1. Update features to match your exact CSV column names
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.kdeplot(
        data=df,
        x=feature,
        hue='Species',   # Capitalized 'S' to match your CSV
        fill=True,       # Fill area under curve
        alpha=0.4,
        ax=axes[i]
    )
    
    # Clean up the title and labels for display
    clean_name = feature.replace("Cm", "").replace("_", " ")
    axes[i].set_title(f'KDE — {clean_name}', fontweight='bold')
    axes[i].set_xlabel(f'{clean_name} (cm)')

plt.suptitle('KDE Plots — Smooth Distribution Comparison', fontsize=15, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('kde_plots.png', bbox_inches='tight', dpi=120)
plt.show()
```

### 5.3 — Box Plot: Spread, Median, and Outliers

```python
# ─── Box Plots for All 4 Features ────────────────────────────────────────────
# Box plots show: median (middle line), IQR (box), whiskers (range), and outliers (dots)

fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(
        data=df,
        x='Species',
        y=feature,
        palette='Set2',
        width=0.5,
        flierprops=dict(marker='o', markersize=6, markerfacecolor='red', alpha=0.6),  # Outlier style
        ax=axes[i]
    )
    axes[i].set_title(f'Box Plot — {feature.replace("_", " ").title()}', fontweight='bold')
    axes[i].set_xlabel('Species')
    axes[i].set_ylabel(f'{feature.replace("_", " ").title()} (cm)')

plt.suptitle('Box Plots — Spread and Outliers per Feature by Species', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('box_plots.png', bbox_inches='tight', dpi=120)
plt.show()
print(" Observation: Setosa has the lowest petal values with very little spread. Versicolor and Virginica have some overlap but differ in median values.")
```

```python
# --- Violin Plot — Box Plot + Distribution Combined ---
# Violin plots combine a box plot with a KDE — richer view of data spread

# 1. Update features to match your exact CSV column names
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

fig, axes = plt.subplots(2, 2, figsize=(13, 9))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.violinplot(
        data=df,
        x='Species',      # Capitalized 'S' to match your CSV
        y=feature,
        palette='Set2',
        inner='quartile', # Show quartile lines inside violin
        ax=axes[i]
    )
    
    # Clean up the title and labels for display
    clean_name = feature.replace("Cm", "").replace("_", " ")
    axes[i].set_title(f'Violin Plot — {clean_name}', fontweight='bold')
    axes[i].set_xlabel('Species')
    axes[i].set_ylabel(f'{clean_name} (cm)')

plt.suptitle('Violin Plots — Distribution + Spread by Species', fontsize=15, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('violin_plots.png', bbox_inches='tight', dpi=120)
plt.show()
```

## Step 6: Correlation Heatmap
A heatmap shows how strongly each pair of numeric features is correlated. Values close to **+1** mean strong positive correlation, **-1** means negative correlation, and **0** means no correlation.

```python
# --- Correlation Heatmap ---
# Compute correlation matrix for numeric columns only

# Option 1: Explicitly drop 'Species' (Capitalized)
# numeric_df = df.drop(columns='Species') 

# Option 2: Automatically select only numeric columns (Recommended)
numeric_df = df.select_dtypes(include=['number'])

corr_matrix = numeric_df.corr().round(2) # Pearson correlation coefficients

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,          # Show correlation values inside cells
    cmap='coolwarm',     # Blue = negative, Red = positive correlation
    fmt='.2f',           # Format as 2 decimal places
    linewidths=0.5,
    square=True,
    vmin=-1, vmax=1      # Fix scale from -1 to +1
)

plt.title('Correlation Heatmap – Feature Relationships', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', bbox_inches='tight', dpi=120)
plt.show()

print(" Observation: Petal length and petal width are highly correlated (0.96).")
```
